```python
import pandas as pd
import plotly.express as px

df = px.data.gapminder()

# Filter data for Europe in years 1952 and 2002
df = df[(df.continent == "Europe") & (df.year.isin([1952, 2002]))]

# Sorting countries by life expectancy in 2002
countries_sorted = df[df.year == 2002].sort_values(by="lifeExp")["country"].unique()
df["country"] = pd.Categorical(df["country"], categories=countries_sorted, ordered=True)

# Reshape data for proper line drawing
df_lines = df.pivot(index="country", columns="year", values="lifeExp").reset_index()
df_lines = df_lines.melt(id_vars=["country"], var_name="year", value_name="lifeExp")
df_lines
```

```python

# Create the line chart connecting 1952 and 2002
fig = px.line(
    df_lines,
    x="lifeExp",
    y="country",
    line_group="country",
    markers=True,
    title="Life Expectancy in Europe: 1952 and 2002",
    labels={"lifeExp": "Life Expectancy", "country": "Country", "year": "Year"},
    height=1000
)
# Update marker colors (Green for 1952, Blue for 2002)
fig.update_traces(marker=dict(size=10))


# Show the figure
fig.show()
```

```python
import pandas as pd

# Sample data: List of project tasks with start and end dates
data = pd.DataFrame({
    "Task": ["Planning", "Design", "Development", "Testing", "Deployment"],
    "Start": ["2024-03-01", "2024-03-10", "2024-03-20", "2024-04-05", "2024-04-15"],
    "End": ["2024-03-09", "2024-03-19", "2024-04-04", "2024-04-14", "2024-04-20"]
})

# Convert date columns to datetime format
data["Start"] = pd.to_datetime(data["Start"])
data["End"] = pd.to_datetime(data["End"])

# Create Gantt chart using plotly
fig = px.timeline(data, x_start="Start", x_end="End", y="Task", title="Project Timeline")

# Reverse order so tasks are listed from top to bottom
fig.update_yaxes(categoryorder="total ascending")

# Show the chart
fig.show()
```

```python
import plotly.express as px
import pandas as pd

# Sample data: Task name, Start date, Duration (in days), and Precedence
data = pd.DataFrame({
    "Task": ["Planning", "Design", "Development", "Testing", "Deployment"],
    "Start": ["2024-03-01", "2024-03-01", "2024-03-01", "2024-03-01", "2024-03-01"],  # Start only for the first task
    "Duration": [9, 10, 15, 9, 5],  # Duration in days
    "Precedence": [None, "Planning", "Design", "Development", "Testing"]  # Dependencies
})

# Convert Start column to datetime format
data["Start"] = pd.to_datetime(data["Start"])

# Calculate Start Dates based on Precedence
for i in range(1, len(data)):  # Start from index 1 since index 0 (Planning) has a fixed start date
    predecessor = data.loc[i, "Precedence"]
    if pd.notna(predecessor):  # If there's a dependent task
        prev_end = data.loc[data["Task"] == predecessor, "Start"].values[0] + pd.to_timedelta(data.loc[data["Task"] == predecessor, "Duration"].values[0], unit="D")
        data.at[i, "Start"] = prev_end  # Assign start after the predecessor ends

# Calculate End Dates
data["End"] = data["Start"] + pd.to_timedelta(data["Duration"], unit="D")

# Create Gantt chart
fig = px.timeline(data, x_start="Start", x_end="End", y="Task", title="Project Timeline with Dependencies")

# Reverse order for better readability
fig.update_yaxes(categoryorder="total ascending")

# Show the chart
fig.show()
```

# Line Plot

### Definition

A line plot is a type of chart used to display data points connected by straight lines. It is mainly used to show trends or changes over a continuous variable, usually time.

### When to use it?

- To show how a variable changes over time

- To visualize trends, patterns, or progressions

- To compare multiple series

### What it shows?

- Trend direction (increasing, decreasing, or constant)

- Rate of change between points

- Comparison between multiple data series (if multiple lines are plotted)

### Important Parameters:

- x → x-axis values

- y → y-axis values

- color → line color

- linestyle → solid, dashed, dotted ('-', '--', ':')

- marker → symbol for each data point ('o', 's', '^')

- label → label for legend

```python
import matplotlib.pyplot as plt
days = [1, 2, 3, 4, 5, 6, 7]
temperature = [22, 24, 21, 23, 25, 26, 24]
plt.plot(days, temperature, color='blue', marker='o', linestyle='-', label='Temp')
plt.title("Daily Temperature Over a Week")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.legend()
plt.show()
```

### Practical Example:

- Plot daily temperature over a month

- Plot stock prices over time

- Plot student scores across different exams

# Scatter Plot

### Definition

A scatter plot displays individual data points as dots on a two-dimensional graph. Each point represents values of two variables (x and y). It is used to observe relationships, correlations, or patterns between the variables.

### When to use it?

- To show relationships between two numeric variables

- To identify trends, clusters, or outliers

- To compare multiple groups by using color or size

### What it shows?

- Correlation: positive, negative, or none

- Outliers or unusual data points

- Clustering or grouping of points

### Important Parameters:

- x → x-axis values

- y → y-axis values

- color / c → color of points

- marker → shape of points ('o', 's', '^')

- s → size of points

- label → for legend

```python
import matplotlib.pyplot as plt
height = [150, 160, 165, 170, 175, 180, 185]
weight = [50, 55, 60, 65, 70, 75, 80]
plt.scatter(height, weight, color='green', marker='o', s=100, label='Students')
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.legend()
plt.show()
```

### Practical Example:

- Plot height vs weight of students

- Plot study hours vs exam scores

- Plot advertising spend vs sales

# Bar Chart

### Definition

A bar chart is used to represent categorical data with rectangular bars. The height (or length) of the bar represents the value of the category.

### When to use it?

- To compare categories

- To show changes over discrete intervals

- To visualize frequency or counts

### What it shows?

- Comparison between categories

- Relative sizes or frequencies

- Trends in categorical data

### Important Parameters:

- x → categories

- y → values

- color → bar color

- width → width of bars

- label → legend label

```python
import matplotlib.pyplot as plt

students = ['Ali', 'Sara', 'Omar', 'Lina']
scores = [85, 90, 78, 92]

plt.bar(students, scores, color='green', label='Scores')
plt.title("Students Scores")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.legend()
plt.show()
```

### Practical Examples:

- Compare sales across products

- Number of students in different departments

- evenue per month

# Histogram

### Definition

A histogram is a graphical representation of the distribution of numerical data. It groups data into bins and shows the frequency of values within each bin.

### When to use it?

- To understand data distribution

- To detect skewness or outliers

- To summarize large datasets

### What it shows?

- Frequency of data values

- Distribution shape (normal, skewed-)

- Outliers or gaps

### Important Parameters:

- data → numeric values

- bins → number of intervals

- color → fill color

- edgecolor → color of bin edges

```python
import matplotlib.pyplot as plt

ages = [22,25,23,24,22,21,26,27,25,24,23,22]
plt.hist(ages, bins=5, color='purple', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### Practical Examples:

- Distribution of grades

- Distribution of ages

- Distribution of income

# Box Plot

### Definition

A box plot (or whisker plot) shows the distribution of data based on five summary statistics: minimum, first quartile, median, third quartile, and maximum. It is useful for identifying outliers.

### When to use it?

- To visualize data spread

- To detect outliers

- To compare distributions across groups

### What it shows?

- Median, quartiles, and range

- Outliers

- Distribution skewness

### Important Parameters:

- data → numeric values

- vert → vertical (True) or horizontal (False)

- patch_artist → fill color of the box

```python
import matplotlib.pyplot as plt

scores = [65,70,75,80,85,90,95,100]
plt.boxplot(scores,vert=True,patch_artist=True)
plt.title("Box Plot of Scores")
plt.show()
```

### Practical Examples:

- Compare salaries across departments

- Detect outliers in exam scores

- Compare test scores across classes

# Pie Chart

### Definition

A pie chart is a circular chart divided into sectors, each representing a proportion of the whole.

### When to use it?

- To show proportions or percentages

- To visualize contribution of categories to total

- For categorical distribution

### What it shows?

- Relative size of each category

- Percentage contribution

- Composition of the whole

### Important Parameters:

- x → values

- labels → category names

- autopct → format for percentages

- startangle → rotation angle

- colors → sector colors

```python
import matplotlib.pyplot as plt

labels = ['Python', 'Java', 'C++', 'R']
sizes = [40, 25, 20, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage")
plt.show()
```

### Practical Examples:

- Budget allocation

- Survey results percentages

# Area Plot

### Definition

An area plot is like a line plot, but the area under the line is filled with color to emphasize magnitude.

### When to use it?

- To show cumulative data

- To visualize trends over time

- To compare multiple series

### What it shows?

- Trend magnitude

- Rate of change

- Comparison of multiple areas

### Important Parameters:

- x → x-axis values

- y → y-axis values

- color → fill color

- alpha → transparency

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]
sales = [100,120,150,130,170]

plt.fill_between(days, sales, color='skyblue', alpha=0.5)
plt.title("Sales Over Time")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
sales_A = [100, 120, 150, 130, 170]
sales_B = [90, 110, 140, 120, 160]
plt.fill_between(days, sales_A, color='skyblue', alpha=0.5, label='Product A')
plt.fill_between(days, sales_B, color='lightgreen', alpha=0.5, label='Product B')
plt.title("Sales Over Time for Two Products")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

### Practical Examples:

- Sales over time

- Website traffic growth

- Population increase

# Stacked Bar Chart

### Definition

A bar chart where multiple series are stacked on top of each other in each category.

### When to use it?

- Show composition of each category

- Compare totals and individual contributions

### What it shows?

- Total values

- Contribution of subcategories

### Important Parameters:

- x → categories

- height → values

- bottom → starting point of the next series

- color → colors for each series

- label → legend label

```python
import matplotlib.pyplot as plt

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
sales_A = [100, 120, 150, 130, 170]
sales_B = [90, 110, 140, 120, 160]

plt.bar(days, sales_A, color='skyblue', label='Product A')
plt.bar(days, sales_B, bottom=sales_A, color='lightgreen', label='Product B')
plt.title("Stacked Sales Over Week")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

# Step Plot

### Definition

A plot that connects data points with horizontal and vertical lines, forming steps instead of straight lines.

### When to use it?

- Show changes that happen at discrete intervals

- Visualize piecewise constant data

### What it shows?

- Sudden changes or discrete transitions

- Step-like trends

### Important Parameters:

- x → x-axis values

- y → y-axis values

- color → line color

- linestyle → style of the line

- label → legend

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]
sales = [100,120,150,130,170]

plt.step(days, sales, color='purple', label='Sales', linestyle='--')
plt.title("Step Plot of Sales")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

The price of a product changes during the day (the price jumps suddenly, not gradually).

```python
import matplotlib.pyplot as plt

time = [1, 2, 3, 4, 5, 6]
price_A = [10, 10, 12, 12, 15, 15]
price_B = [9, 11, 11, 13, 13, 16]

plt.step(time, price_A, where='post', linestyle='-', marker='o', label='Product A')
plt.step(time, price_B, where='post', linestyle='--', marker='s', label='Product B')
plt.title("Step Plot: Product Price Changes Over Time")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.grid(True)

plt.show()
```

### Practical Examples:

- Stock price changes at close of each day

- Water level in a tank measured hourly

- Event occurrences over time

# Stackplot (Area under multiple series)

### Definition

Stackplot shows multiple time series stacked on top of each other. It's like an area plot but for multiple series, showing cumulative contribution.

### When to use it?

- Visualize cumulative data

- Compare composition of multiple series over time

### What it shows?

- Total values over time

- Contribution of each series

### Important Parameters

- x → x-axis values

- y → multiple series

- labels → series names

- colors → fill colors

- alpha → transparency

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]
sales_A = [100,120,150,130,170]
sales_B = [90,110,140,120,160]

plt.stackplot(days, sales_A, sales_B, colors=['skyblue','lightgreen'], labels=['A','B'], alpha=0.5)
plt.title("Stackplot of Sales")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

### Practical Examples

- Population composition over years

- Sales per product category over months

- Energy consumption by source

# Errorbar Plot

### Definition

A line or scatter plot with error bars showing variability or uncertainty in the data.

### When to use it?

- Represent measurement uncertainty

- Show variability in repeated experiments

### What it shows?

- Mean or measured value

- Range of error / confidence intervals

### Important Parameters

- x → x-axis values

- y → y-axis values

- yerr → error values

- fmt → marker style

- color → line color

- label → legend

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]
sales = [100,120,150,130,170]
error = [5, 7, 6, 4, 5]

plt.errorbar(days, sales, yerr=error, fmt='o-', color='red', label='Sales')
plt.title("Sales with Error Bars")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

### Practical Examples

# Violin Plot

### Definition

A violin plot is a visualization that combines features of a box plot and a kernel density plot. It shows the distribution of the data and its probability density.

### When to use it?

- Compare distributions between groups

- Understand the shape of data distribution

- Visualize density of values

### What it shows?

- Distribution shape

- Density of data

- Median and spread of the dataset

### Important Parameters

- data → dataset values

- showmeans → display mean value

- showmedians → display median

- vert → vertical or horizontal plot

```python
import matplotlib.pyplot as plt

math_scores = [65,70,75,80,85,90,95,100]
physics_scores = [60,68,72,78,82,88,91,96]
cs_scores = [70,74,78,83,87,92,97,99]
data = [math_scores, physics_scores, cs_scores]
plt.violinplot(data, showmeans=True, showmedians=True)
plt.xticks([1,2,3], ['Math', 'Physics', 'Computer Science'])
plt.title("Score Distribution Across Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
#plt.grid(True)
plt.show()
```

# Polar Plot

### Definition

A polar plot represents data in polar coordinates where points are determined by an angle and a radius.

### When to use it?

- Visualize circular data

- Analyze directional data

- Represent periodic patterns

### What it shows?

- Angular relationships

- Radial distance patterns

- Cyclic trends

### Important Parameters

- theta → angle values

- r → radius values

- color → line color

- linestyle → line style

```python
import matplotlib.pyplot as plt
import numpy as np
theta = np.linspace(0, 2*np.pi, 100)
r = theta
plt.polar(theta, r)
plt.title("Polar Plot Example")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
theta = np.linspace(0, 2*np.pi, 12)
sensor_A = 5 + 2*np.sin(theta)
sensor_B = 4 + 2*np.cos(theta)
fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')
ax.plot(theta, sensor_A, marker='o', color='blue', label='Sensor A')
ax.plot(theta, sensor_B, marker='s', color='red', linestyle='--', label='Sensor B')
ax.fill(theta, sensor_A, alpha=0.2, color='blue')
ax.set_title("Polar Plot: Sensor Measurements Around a Circle")
ax.legend(loc='upper right')
plt.show()
```

### Practical Examples

- Wind direction analysis

- Radar data visualization

- Circular motion patterns

# Event Plot

### Definition

An event plot visualizes the occurrence of events along a timeline.

### When to use it?

- Display events over time

- Analyze time-based activities

- Visualize event distribution

### What it shows?

- Timing of events

- Event frequency

- Event distribution over time

### Important Parameters

- positions → event locations

- orientation → horizontal or vertical

- colors → event color

- linelengths → length of event lines

```python
import matplotlib.pyplot as plt
events = [1, 2, 3, 5, 8]
plt.eventplot(events)
plt.title("Event Plot Example")
plt.show()
```

```python
import matplotlib.pyplot as plt
user1_events = [1, 3, 5, 7, 9]
user2_events = [2, 4, 6, 8]
user3_events = [1.5, 3.5, 6.5, 9.5]
events = [user1_events, user2_events, user3_events]
plt.eventplot(events, colors=['red', 'blue', 'green'],linelengths=[0.8, 0.8, 0.8],lineoffsets=[1, 2, 3])
plt.yticks([1,2,3], ['User 1', 'User 2', 'User 3'])
plt.title("Event Plot: User Activity Timeline")
plt.xlabel("Time")
plt.ylabel("Users")
plt.grid(True)
plt.show()
```

### Practical Examples

- Earthquake occurrences

- Website user activity

- Machine failure events

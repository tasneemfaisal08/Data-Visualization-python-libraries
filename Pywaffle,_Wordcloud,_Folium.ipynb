# Pywaffle

- is a Python library used to create Waffle Charts for data visualization.

- A **Waffle Chart** represents data using a grid of squares where each square represents a proportion of the dataset.

- PyWaffle is commonly used to show percentage distributions in an - easy-to-understand visual format.

- Core feature: Waffle chart visualization

- Used in: Data storytelling, dashboards, reports, and presentations

# Installation

```python
pip install pywaffle
```

# Importing

```python
from pywaffle import Waffle
import matplotlib.pyplot as plt
```

```python
data = {'Python': 50,'R': 20,'SQL': 15,'Excel': 15}
fig = plt.figure(FigureClass=Waffle,rows=5,values=data,title={'label': 'Skills Used in Data Science', 'loc': 'center'},
    legend={'loc': 'lower left', 'bbox_to_anchor': (1, 0)})
plt.show()
```

**What should you understand from this chart?**

A Waffle Chart represents proportions using squares.

**Example interpretation:**

Skill |	Percentage
------|---------
Python |	50%
R	| 20%
SQL |	15%
Excel	| 15%

**If the chart has 100 squares:**

- 50 squares represent Python

- 20 squares represent R

- 15 squares represent SQL

- 15 squares represent Excel

**Conclusion**

1. Python is the most commonly used skill.

2. The other skills are less frequently used.


Waffle charts are useful when you want to show how each category contributes to the whole dataset.

# WordCloud

- is a Python library used to generate word cloud visualizations from text data.

- A Word Cloud displays words in different sizes based on their frequency or importance in a text.

- The more frequent a word appears in the text, the larger it appears in the visualization.

- Core feature: Text frequency visualization

- Used in: Text analysis, Natural Language Processing (NLP), social media     analysis, and text data exploration

# Installation

```python
pip install wordcloud
```

# Importing

```python
from wordcloud import WordCloud
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = """
data science machine learning python data analysis
python statistics machine learning visualization
data science python ai machine learning
"""

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

**What should you understand from this chart?**

A Word Cloud shows word frequency in a text.

**The rule is:**

- Larger words → appear more frequently

- Smaller words → appear less frequently

**For example:**

If **"python"** appears as the largest word, it means it is the most frequent word in the text.


**Conclusion:**

The most prominent words represent the main topics of the text.




**Word clouds are commonly used in:**

1. Text analysis

2. Natural Language Processing (NLP)

3. Social media analysis

4. Review analysis

**Example interpretation:**

If a restaurant review WordCloud contains large words such as:

- delicious

- good

- service

➡ This suggests that the restaurant reviews are mostly positive.

# Folium

- is a Python library used for creating **interactive maps.**

- It allows users to visualize geographical data by adding markers, heatmaps, and layers on maps.

- Folium is built on top of Leaflet.js, a popular JavaScript library for interactive maps.

- Core feature: Interactive geographic visualization

- Used in: Geospatial analysis, location-based data visualization, mapping projects, and dashboards

# Installation

```python
pip install folium
```

# Importing

```python
import folium
```

```python
import folium

# Create map centered at Cairo
map = folium.Map(location=[30.0444, 31.2357], zoom_start=12)
# Add marker
folium.Marker(
    location=[30.0444, 31.2357],
    popup="Cairo",
    tooltip="Click here"
).add_to(map)
```

```python
map
```

**What should you understand from this chart?**

Folium is used to **visualize geographical** data on interactive maps.

**In this example:*

📍 A marker is placed on Cairo.

This type of visualization can be used for:

- Customer locations

- Accident locations

- Store branches

- GPS tracking data

**Example analysis**

If a map shows many accident markers, and most of them are concentrated in one area:

➡ The conclusion is that this area is a high-risk zone for accidents.

This information could help authorities improve traffic safety in that area.

# Seaborn

Seaborn is a high-level data visualization library built on top of Matplotlib.
It provides a simpler interface to create attractive and informative statistical graphics.

- Designed for statistical data visualization

- Built on Matplotlib, works well with Pandas DataFrames

- Core feature: high-level plotting functions (scatter, line, bar, violin, heatmap…)

- Used in: Data Science, Machine Learning, EDA, Statistical Analysis, Reporting

- Integrates easily with NumPy and Pandas

# Relarional Plots

Show the relationship between two numerical variables and reveal trends or patterns.

## Line Plot

### Definition

A line plot displays trends over time or ordered data.

### When to use it?

- Analyze time series

- Track changes over time

- Compare multiple trends

### What it shows?

- Trend direction

- Changes over time

- Comparison between groups

### Important Parameters

- x → x variable

- y → y variable

- hue → grouping variable

- style → line style

- markers → show markers

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("flights")
sns.lineplot(x="year", y="passengers", data=data)
plt.title("Passenger Trend Over Time")
plt.show()
```

### Practical Examples

- Stock prices over time

- Website traffic

- Sales growth

## Scatter Plot

### Definition

A scatter plot shows the relationship between two numerical variables using points.

### When to use it?

- Analyze relationship between variables

- Detect clusters or patterns

- Identify outliers

### What it shows?

- Correlation between variables

- Distribution of data points

- Clusters and trends

### Important Parameters

- x → x variable

### Practical Examples

## Relplot

### Definition

Relplot is a figure-level function in Seaborn used to plot relationships between two numerical variables. It can create scatter plots or line plots and supports faceting (multiple plots by category).

### When to use it?

- Explore the relationship between two numerical variables

- Compare trends across categories

- Create multiple subplots (facets) for grouped data

### What it shows?

- Correlation or trend between variables

- Differences between categories

- Patterns across subsets of data

### Important Parameters

- x → variable for the x-axis

- y → variable for the y-axis

- hue → grouping variable (assigns different colors to groups)

- size → variable that changes the size of the markers

- style → variable that changes the marker style

- kind → "scatter" (default) or "line"

- col → variable to create separate plots as columns

- row → variable to create separate plots as rows

- palette → color palette for different groups

- markers → marker types to use for different style levels

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
sns.relplot(x="total_bill",y="tip",hue="sex",style="time",size="size",kind="scatter",data=data)
plt.title("Relational Plot: Tips vs Total Bill")
plt.show()
```

create separate plots for lunch and dinner

```python
sns.relplot(x="day",y="total_bill",hue="sex",kind="line",col="time",data=data)
plt.show()
```

### Practical Examples

- Total bill vs tip analysis in restaurants (by gender, time)

- Sales vs advertising spend across regions

- Exam scores vs study hours by class

## Boxen Plot

### Definition

Boxen Plot (a.k.a Letter Value Plot) is an enhanced version of a box plot that shows more quantiles, making it better for large datasets or datasets with long tails.

### When to use it?

- Visualize the distribution of data for large datasets

- Detect outliers and skewed distributions

- Compare multiple categories

### What it shows?

- Median and quantiles

- Tail distributions

- Outliers

### Important Parameters

- x → categorical variable

- y → numeric variable

- hue → grouping variable

- palette → color palette

- scale → scale of boxes ("linear", "exponential")

- orient → "v" for vertical, "h" for horizontal

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
sns.boxenplot(x="day", y="total_bill", hue="sex", data=data, palette="Set2")
plt.title("Boxen Plot: Total Bill per Day by Gender")
plt.show()
```

### Practical Examples

- Salary distribution in a company (large dataset)

- House prices per region

- Exam scores with many students

## Catplot

### Definition

catplot is a figure-level function for categorical plots; it can create bar, box, violin, strip, swarm, point plots in a single interface.

### When to use it?

- Compare categorical data easily

- Combine multiple plot types in one interface

- Facet plots across rows/columns

### What it shows?

- Summary statistics (mean, median)

- Distribution of categorical variables

- Differences between groups

### Important Parameters

- x → categorical variable

- y → numeric variable

- hue → subgroup variable

- kind → "bar", "box", "violin", "strip", "swarm", "point"

- col → facet by column

- row → facet by row

- palette → color scheme

- height → height of each facet

- aspect → width/height ratio of each facet

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", hue="sex", kind="violin", data=data, palette="Pastel1", height=5, aspect=1.5)
plt.title("Catplot (Violin) of Total Bill per Day by Gender")
plt.show()
```

### Practical Examples

- Average sales per product category (bar plot)

- Exam score distribution per class (violin or box plot)

- Customer tips by day and gender

# Matrix Plots

Visualize relationships or values in a matrix form (correlations or intensity).

## Heatmap

### Definition

A heatmap visualizes matrix values using colors.

### When to use it?

- Analyze correlations

- Identify patterns in matrices

### What it shows?

- Correlation strength

- Data intensity

### Important Parameters

- data → matrix

- annot → show values

- cmap → color map

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### Practical Examples

- Correlation matrix

- Feature relationships in ML

## Clustermap

### Definition

A Clustermap is a matrix plot that combines a heatmap with hierarchical clustering.
It automatically reorders rows and columns based on similarity, helping to identify patterns and clusters in data.

### When to use it?

- Analyze correlation or similarity matrices

- Detect clusters in large datasets

- Visualize patterns across multiple variables simultaneously

### What it shows?

- Heatmap of values

- Clustering dendrograms for rows and columns

- Patterns and relationships in data

### Important Parameters

- data → 2D dataset (matrix or DataFrame)

- method → clustering method ("single", "complete", "average")

- metric → distance metric ("euclidean", "correlation")

- standard_scale → 0 for column-wise, 1 for row-wise standardization

- figsize → size of the figure

- cmap → color map for heatmap

- row_cluster / col_cluster → True/False to cluster rows/columns

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
data = sns.load_dataset("flights")
pivot_data = data.pivot(index="month", columns="year", values="passengers")
sns.clustermap(pivot_data, cmap="coolwarm", standard_scale=1)
plt.title("Clustermap: Passengers by Month and Year")
plt.show()
```

# Grid Plots (Multi-plot systems)

Create multiple small plots together to compare data across

## FacetGrid

### Definition

FacetGrid is a grid plotting system that lets you create multiple subplots based on categories or other variables. It’s flexible and works with any axes-level Seaborn plot.

### When to use it?

- Compare distributions or relationships across categories

- Create multiple subplots automatically

- Visualize patterns in subsets of data

### What it shows?

- Multiple plots organized in rows and columns

- Differences between categories

- Trends across subsets

### Important Parameters

- data → dataset (DataFrame)

- row → variable to create rows of subplots

- col → variable to create columns of subplots

- hue → variable to color points/groups

- palette → color palette

- height → height of each subplot

- aspect → width/height ratio

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
g = sns.FacetGrid(data, col="time", row="sex", height=3, aspect=1)
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```

### Practical Examples

- Compare sales vs marketing spend for different regions and months

- Student scores vs study hours by gender and class

## PairGrid

### Definition

PairGrid creates a grid of plots to show pairwise relationships between multiple variables. You can customize upper, lower, and diagonal plots separately.

### When to use it?

- Explore pairwise relationships in a dataset

- Visualize correlation patterns

- Combine scatter, histograms, and KDEs in one grid

### What it shows?

- Scatter plots between pairs of variables

- Histograms or KDEs on the diagonal

- Patterns across multiple numerical variables

### Important Parameters

- data → dataset (DataFrame)

- hue → variable to color points/groups

- palette → color palette

- height → height of each subplot

- aspect → width/height ratio

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("iris")
g = sns.PairGrid(data, hue="species", height=2.5)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
g.add_legend()
plt.show()
```

### Practical Examples

- Analyze relationships between multiple features of flowers (iris dataset)

- Compare financial indicators across companies

## JointGrid

### Definition

JointGrid is a grid for plotting a bivariate relationship between two variables, with marginal plots on the top and right for each variable.

### When to use it?

- Explore relationship between two variables

- Show distributions of each variable along with scatter or regression

- Customize central and marginal plots separately

### What it shows?

- Central plot: scatter, regression, KDE, hexbin

- Marginal plots: histograms, KDEs for each variable

- Combination of joint and marginal distributions

### Important Parameters

- x → numeric variable for x-axis

- y → numeric variable for y-axis

- data → dataset (DataFrame)

- hue → grouping variable

- height → size of the square grid

- ratio → ratio of central plot to marginals

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
g = sns.JointGrid(x="total_bill", y="tip", data=data, height=6)
g.plot(sns.scatterplot, sns.histplot)
plt.show()
```

### Practical Examples

- Tips vs total bill analysis (scatter + histograms)

- Exam scores vs study hours with marginal distributions

- Height vs weight of students
